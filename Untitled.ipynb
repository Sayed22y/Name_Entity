```python
import spacy 
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp("Im going to finish our course tonight")
```

```python
for word in doc :
    print(word.lemma_)
```

```python
for word in doc:
    print(word.text, word.dep_, word.head.pos_)
```

```python
text = 'John Luo loves Ali Hassan when driving in the highway. Ali Hassan said thath Apple Inc. from Calefornia will buy Google in may 2025'
doc = nlp(text)
```

```python
for ent in doc.ents:
    print(ent.text, ent.label_, ent.start_char, ent.end_char)
```

```python
from spacy import displacy
```

```python
doc = nlp('To be or not to be this is qusetion')
displacy.render(doc, style='dep', jupyter=True, options={'distance' : 120})
```

```python
doc = nlp("I'm eating fish, My friend Ahmed doesn't like fish. I'm going to leave my work. Next week I will be at Google")
displacy.render(doc, style='ent', jupyter=True)
```

```python
doc = nlp("my dog still eating banana")

for word in doc:
    for word2 in doc:
        print(word, word2, word.similarity(word2))
```
